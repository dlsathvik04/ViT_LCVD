```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the quantization results
data = {
    "Level": ["FP32 (Original)", "FP16", "INT8"],
    "Accuracy": [0.9896, 0.9896, 0.9896],
    "Size (MB)": [327.42, 163.75, 84.44]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Display the table
print("Quantization Results:")
print(df.to_string(index=False, line_width=40))

# Plot Accuracy vs Size
plt.figure(figsize=(8, 6))
plt.plot(df["Size (MB)"], df["Accuracy"], color='blue')
plt.scatter(df["Size (MB)"], df["Accuracy"], color='blue', s=100, label='Quantization Levels')
for i, level in enumerate(df["Level"]):
    plt.annotate(level, (df["Size (MB)"][i], df["Accuracy"][i]), xytext=(5, 5), 
                 textcoords='offset points', fontsize=10)
plt.xlabel("Model Size (MB)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Model Size Across Quantization Levels")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Save the plot
plt.savefig("results/quantization_plot.png")
print("\nPlot saved as 'quantization_plot.png'")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from the classification reports (macro and weighted averages only)
data = {
    "Model": ["Distillation", "Direct Training", "Parent/Original"],
    "Macro Precision": [0.99, 0.98, 0.99],
    "Macro Recall": [0.98, 0.97, 0.99],
    "Macro F1": [0.98, 0.98, 0.99],
    "Weighted Precision": [0.99, 0.98, 0.99],
    "Weighted Recall": [0.99, 0.98, 0.99],
    "Weighted F1": [0.99, 0.98, 0.99],
    "Accuracy": [0.99, 0.98, 0.99]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Display the summary table
print("Classification Report Summary:")
print(df.to_string(index=False))

# Set up the bar plot
metrics = ["Macro Precision", "Macro Recall", "Macro F1", 
           "Weighted Precision", "Weighted Recall", "Weighted F1"]
models = df["Model"]
n_models = len(models)
n_metrics = len(metrics)

# Prepare data for plotting
bar_width = 0.25
index = np.arange(n_metrics)

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars for each model
for i, model in enumerate(models):
    values = df.iloc[i][metrics].values
    ax.bar(index + i * bar_width, values, bar_width, label=model)

# Customize the plot
ax.set_xlabel("Metrics")
ax.set_ylabel("Score")
ax.set_title("Comparison of Classification Metrics Across Models")
ax.set_xticks(index + bar_width * (n_models - 1) / 2)
ax.set_xticklabels(metrics, rotation=45, ha="right")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)
plt.ylim(0.95, 1.0)  # Adjust y-axis to zoom in on differences
plt.tight_layout()

# Save the plot
plt.savefig("results/distillation_comparison.png")
print("\nPlot saved as 'distillation_comparison.png'")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from the classification reports (macro averages and accuracy)
data = {
    "Model": ["Distillation", "Direct Training", "Parent/Original"],
    "Macro Precision": [0.99, 0.98, 0.99],
    "Macro Recall": [0.98, 0.97, 0.99],
    "Macro F1": [0.98, 0.98, 0.99],
    "Accuracy": [0.99, 0.98, 0.99]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Display the summary table
print("Macro Average and Accuracy Comparison:")
print(df.to_string(index=False))

# Set up the bar plot
metrics = ["Macro Precision", "Macro Recall", "Macro F1", "Accuracy"]
models = df["Model"]
n_models = len(models)
n_metrics = len(metrics)

# Prepare data for plotting
bar_width = 0.2
index = np.arange(n_metrics)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for each model
for i, model in enumerate(models):
    values = df.iloc[i][metrics].values
    ax.bar(index + i * bar_width, values, bar_width, label=model)

# Customize the plot
ax.set_xlabel("Metrics")
ax.set_ylabel("Score")
ax.set_title("Macro Average and Accuracy Comparison Across Models")
ax.set_xticks(index + bar_width * (n_models - 1) / 2)
ax.set_xticklabels(metrics, rotation=45, ha="right")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)
plt.ylim(0.95, 1.0)  # Zoom in to highlight differences
plt.tight_layout()

# Save the plot
plt.savefig("results/macro_accuracy_distillation_comparison.png")
print("\nPlot saved as 'macro_accuracy_distillation_comparison.png'")

# Show the plot
plt.show()
```
